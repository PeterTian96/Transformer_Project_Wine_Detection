```python
import pandas as pd
from datasets import Dataset, DatasetDict
from datasets import load_dataset

data = pd.read_excel("winereview_data.xlsx")
data.head()
```

```python
data.shape
```

```python
# change n to detereming selected data size
n = len(data)
sample = data.sample(n=n, random_state=42)

train_size = int(n*0.7)
val_size = int(n*0.15)
test_size = int(n*0.15)
train_data = sample[:train_size]
val_data = sample[train_size:train_size+val_size]
test_data = sample[train_size+val_size:train_size+val_size+test_size]
```

```python
train_data = train_data.astype(str)
val_data = val_data.astype(str)
test_data = test_data.astype(str)
```

```python
train_dataset = Dataset.from_pandas(train_data)
val_dataset = Dataset.from_pandas(val_data)
test_dataset = Dataset.from_pandas(test_data)

dataset = DatasetDict({
    "train": train_dataset,
    "validation": val_dataset,
    "test": test_dataset
})

for split in dataset.keys():
    print(f"{split}: {len(dataset[split])}")
```

```python
def show_samples(dataset,num_samples = 3,seed = 40):
  sample = dataset["train"].shuffle(seed=seed).select(range(num_samples))
  for example in sample:
    print(f"\n'Title:{example['variety']}")
    print(f"'Review:{example['description']}")

show_samples(dataset)
```

```python
dataset
```

```python
from transformers import AutoTokenizer

model_checkpoint = "google/mt5-small"
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)
```

```python
max_input_length = 512
max_target_length = 30

def preprocess_function(examples):
  model_inputs = tokenizer(
      examples["description"],max_length=max_input_length,truncation=True
  )
  with tokenizer.as_target_tokenizer():
    labels = tokenizer(
        examples["variety"],max_length=max_target_length,truncation=True
    )
  
  model_inputs["labels"] = labels["input_ids"]
  return model_inputs
```

```python
tokenized_datasets = dataset.map(preprocess_function,batched=True)
```

```python
tokenized_datasets
```

```python
from datasets import load_metric

rouge_score = load_metric("rouge")
```

```python
from transformers import AutoModelForSeq2SeqLM
model = AutoModelForSeq2SeqLM.from_pretrained(model_checkpoint)
```

```python
from transformers import Seq2SeqTrainingArguments

batch_size = 8
num_train_epochs = 5

logging_steps = len(tokenized_datasets["train"])// batch_size
model_name = model_checkpoint.split("/")[-1]

args = Seq2SeqTrainingArguments(
    output_dir = f"{model_name}-wine-recommendation-mt5model_0407",
    evaluation_strategy="epoch",
    learning_rate=5.6e-5,
    per_device_train_batch_size=batch_size,
    per_device_eval_batch_size=batch_size,
    weight_decay=0.01,
    save_total_limit=1,
    num_train_epochs=num_train_epochs,
    predict_with_generate=True,
    logging_steps=logging_steps,
    save_strategy='epoch'
)
```

```python
import numpy as np

def compute_metrics(eval_pred):
  predictions,labels=eval_pred
  result=rouge_score.compute(
      predictions=predictions,references=labels,use_stemmer=True
  )
  result= {key: value.mid.fmeasure * 100 for key,value in result.items()}
  return {k:round(v,4) for k,v in result.items()}
```

```python
from transformers import DataCollatorForSeq2Seq

data_collator=DataCollatorForSeq2Seq(tokenizer,model=model)

tokenized_datasets=tokenized_datasets.remove_columns(
    dataset["train"].column_names
)
```

```python
features=[tokenized_datasets["train"][i] for i in range(2)]
data_collator(features)
```

```python
from transformers import Seq2SeqTrainer

trainer = Seq2SeqTrainer(
    model,
    args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["validation"],
    data_collator=data_collator,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics
)
```

```python
trainer.train()
```

```python
from transformers import pipeline

summarizer=pipeline("summarization",model='./mt5-small-wine-recommendation-mt5model_0407/checkpoint-56865')
```

```python
def print_summary(idx):
  description=dataset["test"][idx]["description"]
  variety=dataset["test"][idx]["variety"]
  summary=summarizer(dataset["test"][idx]["description"])[0]["summary_text"]
  print(f"'>>>Description:{description}'")
  print(f"\n'>>>Variety:{variety}'")
  print(f"\n'>>>Summary:{summary}'")
```

```python
print_summary(1)
```

```python
# Evaluate the model
evaluation_results = trainer.evaluate(eval_dataset=tokenized_datasets["validation"])

# Access and print the metrics from the evaluation_results dictionary
for metric, value in evaluation_results.items():
    print(f"{metric}: {value:.4f}")
```
